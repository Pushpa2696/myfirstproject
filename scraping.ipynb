```python


import os
import pandas as pd

# ✅ Define Paths
csv_folder = r"C:\Users\vishn\OneDrive\Desktop\Project1"
merged_csv_path = os.path.join(csv_folder, "merged_movies.csv")

# ✅ Get All CSV Files
csv_files = [file for file in os.listdir(csv_folder) if file.endswith(".csv")]

# ✅ List to Store Data
all_data = []

# ✅ Merge CSVs
for file in csv_files:
    file_path = os.path.join(csv_folder, file)

    try:
        if os.path.getsize(file_path) == 0:
            print(f"⚠️ Skipping empty file: {file}")
            continue

        df = pd.read_csv(file_path, encoding='utf-8', sep=None, engine='python')

        # ✅ Extract Genre from File Name
        df["Genre"] = file.replace(".csv", "")

        # ✅ Ensure Consistent Column Order (move "Genre" to last)
        df = df[[col for col in df.columns if col != "Genre"] + ["Genre"]]

        all_data.append(df)

    except Exception as e:
        print(f"❌ Error processing {file}: {e}")

# ✅ Merge & Save Final CSV
if all_data:
    final_df = pd.concat(all_data, ignore_index=True)
    final_df.to_csv(merged_csv_path, index=False)
    print(f"✅ Merged CSV saved to: {merged_csv_path}")
else:
    print("❌ No valid CSV files found.")
```

```python


import pandas as pd

#  Load the cleaned CSV file

df = pd.read_csv(r"C:\Users\vishn\OneDrive\Desktop\Project1\merged_movies_cleaned.csv")

#  Convert Votes Column (K → 1000, M → 1000000)
def convert_votes(votes):
    if pd.isna(votes):
        return None
    votes = str(votes).upper().strip()
    if "M" in votes:
        return int(float(votes.replace("M", "")) * 1_000_000)
    if "K" in votes:
        return int(float(votes.replace("K", "")) * 1_000)
    return int(votes) if votes.isdigit() else None

df["Votes"] = df["Votes"].apply(convert_votes)

#  Clean Duration Column (Remove M)
df["Duration(minutes)"] = df["Duration(minutes)"].astype(str).str.replace("m", "").astype(float)

#  Save the cleaned data back to a new CSV file
output_file = r"C:\Users\vishn\OneDrive\Desktop\Project1\final_movies_cleaned.csv"
df.to_csv(output_file, index=False)

print(f" Cleaned data saved to: {output_file}")
```
